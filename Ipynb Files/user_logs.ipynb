```python
import numpy as np #Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
currentdir = os.getcwd() #Get current Working Directory
currentdir = currentdir[:len(currentdir)-11] # Replacing The Current directory
currentdir += "Datasets/server-logs/" #Directory Of Dataset Used
file = currentdir + "processed_logs.csv" #File name
df = pd.read_csv(file) #Reading The File
df.head() #Displaying the file
```

```python
df.replace(to_replace='-',value=np.nan,inplace=True) # Replacing '-' with NaN
df.dropna(how='any',inplace = True) #Dropping any row with NaN value
df.head() #output
```
