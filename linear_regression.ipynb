```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("test.csv")
df.replace([np.inf,-np.inf],np.nan,inplace=True)
df=df.dropna()
df.isnull()
```

```python
corr_data = df.corr()
```

```python
corr_data
```

```python
plt.figure(figsize=(40,20))
sns.heatmap(corr_data,vmin=-1,vmax=1,annot=True,cmap='RdBu')
```

```python
df.dtypes
```

```python
# Dependant and independant variable
x=df.drop(columns='Variance')
y=df['Variance']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=20,shuffle=True)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred_train = lr.predict(x_train)
```

```python
plt.scatter(y_train,y_pred_train)
plt.xlabel("Actual Variance")
plt.ylabel("Variance")
plt.title("Training Data")
plt.show()
```

```python
y_pred_test = lr.predict(x_test)
```

```python
plt.scatter(y_test,y_pred_test)
plt.xlabel("Actual Variance")
plt.ylabel("Variance")
plt.title("Testing Data")
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(y_train,y_pred_train)
```

```python
r2_score(y_test,y_pred_test)
```
